```python
# There are two tables with user activities. 
# The google_gmail_emails table contains information about emails being sent to users. 
# Each row in that table represents a message with a unique identifier in the id field. 
# The google_fit_location table contains user activity logs from the Google Fit app.
#Find the correlation between the number of emails received and the total exercise per day. 
# The total exercise per day is calculated by counting the number of user sessions per day.

gme = google_gmail_emails
gme = gme.groupby(['to_user', 'day'])['id'].nunique().to_frame('num_emails').reset_index()

gfl = google_fit_location
gfl = gfl[['user_id', 'session_id', 'day']]
gfl = gfl.groupby(['user_id', 'day'])['session_id'].nunique().to_frame('total_excercise').reset_index()

merged = pd.merge(gme, gfl, left_on=['to_user', 'day'], right_on=['user_id', 'day'], how='outer').fillna(0)

corr = merged['num_emails'].corr(merged['total_excercise'])
corr
```

```python
# Find how the number of drivers that have churned changed in each year compared to the previous one. 
# Output the year (specifically, you can use the year the driver left Lyft) along with the corresponding number 
# of churns in that year, the number of churns in the previous year, and an indication on whether the number has 
# been increased (output the value 'increase'), decreased (output the value 'decrease') 
# or stayed the same (output the value 'no change').

lyft_drivers = lyft_drivers[~lyft_drivers['end_date'].isnull()]
lyft_drivers['end_year'] = lyft_drivers['end_date'].dt.year
driver_counts = lyft_drivers.groupby('end_year')['index'].count().to_frame('churn_count').reset_index()
driver_counts['previous_year'] = driver_counts['end_year'] - 1
driver_counts['prev_churn_count'] = driver_counts['churn_count'].shift(1).fillna(0)
driver_counts['status'] = 'tbd'
for i in range(1, len(driver_counts)):
    if driver_counts.loc[i, 'churn_count'] - driver_counts.loc[i-1, 'churn_count'] == 0:
        driver_counts.loc[i, 'status'] = 'no change'
    elif driver_counts.loc[i, 'churn_count'] - driver_counts.loc[i-1, 'churn_count'] > 0:
        driver_counts.loc[i, 'status'] = 'increase'
    else:
        driver_counts.loc[i, 'status'] = 'decrease'
driver_counts
```
